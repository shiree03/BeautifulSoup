```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.dineout.co.in/bangalore-restaurants?city_name=bangalore&limit=21&start=0&cityId=2&listing=1&tag%5B%5D=Buffet&showAvailableTicket=0')
```

```python
page
#200=we can proceed with webscrapping
#501,500,503=cannot to the web scrapping
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
first_title=soup.find('div',class_="restnt-info cursor")
first_title
```

```python
first_title.text
```

```python
loc=soup.find('div',class_="restnt-loc ellipsis")
loc
```

```python
loc.text
```

```python
price=soup.find("span",class_="double-line-ellipsis")
price.text.split()
```

```python
price.text.replace('₹','')
```

```python
price.text.split()[1:2]
```

```python
titles=[]
for i in soup.find_all('div',class_="restnt-info cursor"):
    titles.append(i.text)
titles
```

```python
location=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
location
```

```python
price=[]
for i in soup.find_all("span",class_="double-line-ellipsis"):
    price.append(i.text.replace('₹',''))
price
```

```python
images=[]
for i in soup.find_all("img",class_="no-img"):
    images.append(i['data-src'])
images
```

```python
print(len(titles),len(location),len(price),len(images))
```

```python
import pandas as pd
df=pd.DataFrame({'Titles':titles,'Location':location,'Price':price,'Images':images})
df
```
